```
from sympy import symbols, Eq, solve

x, y = symbols('x y')
eq1 = Eq(3*x + 5*y, 380)
eq2 = Eq(x + y, 120)

ans = solve((eq1, eq2), (x, y))
a = ans[x]
b = ans[y]
print(f"資淺業務員須外出天數 = {a} 天")
print(f"資深業務員須外出天數 = {b} 天")
print(ans[x] > ans[y])
```

```
from sympy import symbols, Eq, solve
import matplotlib.pyplot as plt
import numpy as np

x, y = symbols('x y')
ans = solve((Eq(3*x + 8*y, 400), Eq(x + y, 120)), (x, y))
a, b = ans[x], ans[y]

X = np.linspace(0, 150, 200)
plt.plot(X, (400 - 3*X)/8, 'r', label='3x + 8y = 400')
plt.plot(X, 120 - X, 'b', label='x + y = 120')
plt.scatter(a, b, c='g', s=60, label=f' ({a}, {b})')

plt.xlim(0, 120)
plt.ylim(0, 60)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()

print(f"資淺業務員須外出天數 = {a} 天")
print(f"資深業務員須外出天數 = {b} 天")
print(a > b)
```
